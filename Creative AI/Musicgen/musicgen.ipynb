```python
import baseten
import os
from dotenv import load_dotenv
load_dotenv()

baseten.login(os.environ["MY_BASETEN_API_KEY"])
#baseten.login("wtk8oz18.CLm4HWr15FGcbgRjXtl3mLtowHNrnUDF")

model = baseten.deployed_model_id('VBl31Vq')
model.predict({'prompts': ['rap', 'energetic EDM', 'sad jazz'], 'duration': 8})
```

```python
import base64
model_output = model.predict({'prompts': ['rap', 'chill house', 'Afrobeat'], 'duration': 8})
for idx, clip in enumerate(model_output["data"]):
  with open(f"clip_{idx}.wav", "wb") as f:
    f.write(base64.b64decode(clip))
```
