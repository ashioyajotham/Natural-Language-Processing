```python
import pandas as pd
import requests
import re
import os
from bs4 import BeautifulSoup as bs
import json

# Get the list of books
url = 'https://www.biblestudytools.com/kjv/'

# Get the page
page = requests.get(url)

# Create a BeautifulSoup object
soup = bs(page.text, 'html.parser')
print(soup)
```

```python
# Get the list of books
# <a class="text-center text-lg bg-gray-100 inline-block hover:bg-yellow-400 hover:text-yellow-800 text-black p-5 border w-full rounded-md" href="https://www.biblestudytools.com/kjv/matthew/">Matthew</a>

books =  soup.find_all('a', class_="text-center text-lg bg-gray-100 inline-block hover:bg-yellow-400 hover:text-yellow-800 text-black p-5 border w-full rounded-md")
'm'
# Create a list of book names
book_names = []
for book in books:
    book_names.append(book.text)
print(book_names)
```

```python
# Create a list of book links
book_links = []
for book in books:
    book_links.append(book['href'])
print(book_links)
```

```python
# Create a list of book abbreviations
book_abbrs = []
for book in books:
    book_abbrs.append(book['href'].split('/')[-2])
print(book_abbrs)
```

```python
# Generate a list of tuples
book_list = list(zip(book_names, book_links, book_abbrs))
print(book_list)
```
